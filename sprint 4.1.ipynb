## Ejercicio 1. Consumir una API

Selecciona una API pública y extrae datos utilizando Python.

```python
import requests
from pandas import DataFrame
url = 'https://dragonball-api.com/api/characters'
data = requests.get(url)
if data.status_code == 200:
    data = data.json()
data = data['items']
df = DataFrame(data, columns=['name','description'])
print(df)
```

## Ejercicio 2. Obtener datos con Web Scraping

Selecciona un sitio web y extrae datos utilizando la técnica de Web Scraping en Python.

```python
import requests
from bs4 import BeautifulSoup

page = requests.get('https://realpython.github.io/fake-jobs/')
soup = BeautifulSoup(page.text, 'html.parser')

results = soup.find(id="ResultsContainer")
jobs = results.find_all('div', class_='card-content')

for job in jobs:
    title = job.find(class_="title")
    company = job.find(class_="company")
    location = job.find(class_="location")
    print('Job: ',title.text)
    print('Company: ',company.text)
    print('Location: ',location.text)
```
